```python
# 1. Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#highest performance score
highestScore = max(employees.values())

#all employees with the same highest score
topPerformers = []
for name,score in employees.items():
  if score == highestScore:
    topPerformers.append(name)

#result
print(f"Top Performers Eligible for Bonus: {', '.join(topPerformers)} (Score: {highestScore})")
```

```python
# 2.Search Query Keyword Analysis
query = "Buy mobile phone buy phone online"

#Convert to lowercase
query = query.lower()

#Split into words
words = query.split()

#Count frequency
wordCnt = {}

for word in words:
    wordCnt[word] = wordCnt.get(word, 0) + 1

#Keep only words which are appear more than once
repeatedWords = {}

for word, count in wordCnt.items():
    if count > 1:
        repeatedWords[word] = count

#result
print(repeatedWords)
```

```python
# 3.Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
validReadings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        validReadings.append((hour, value))

#result
print("Valid Sensor Readings (Hour, Value):")
print(validReadings)
```

```python
# 4.Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domainCnt = {}
#Count users
for email in emails:
    domain = email.split("@")[1]  # Separate domain name
    domainCnt[domain] = domainCnt.get(domain, 0) + 1

#Calculate percentage
total_users = len(emails)

for domain, count in domainCnt.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# 5.Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

#Average sales
average_sales = sum(sales) / len(sales)

#Threshold (30% above average)
threshold = average_sales * 1.30

#Days
for i in range(len(sales)):
    if sales[i] > threshold:
        print(f"Day {i+1}: {sales[i]}")
```

```python
# 6.Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
idCnt = {}

#Count occurrences of each user ID
for uid in user_ids:
    idCnt[uid] = idCnt.get(uid, 0) + 1

#Show duplicates
for uid, count in idCnt.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
