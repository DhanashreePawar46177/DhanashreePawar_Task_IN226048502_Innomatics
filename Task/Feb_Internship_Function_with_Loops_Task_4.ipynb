```python
# 1.Mobile Recharge Validation System
def validateRecharge(amount):
    validPlans = [199, 299, 399, 599]

    #Minimum recharge amount
    if amount < 50:
        print("Recharge amount is less than 50")
        return False

    #Check plan is valid or not
    if amount not in validPlans:
        print("Invalid plan selected")
        print("Available plans are :", validPlans)
        return False

    #all conditions are satisfied
    return True


#retry option
while True:
    userAmount = int(input("Enter recharge amount :"))

    if validateRecharge(userAmount):
        print("Recharge Successful")
        break
    else:
        print("Please try again.\n")
```

```python
# 2.Inventory Reorder Alert System
def checkStock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product} = Reorder Alert (Stock : {quantity})")
        else:
            print(f"{product} = Stock OK (Stock : {quantity})")


#Dictionary of products and quantities
products = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 25,
    "Salt": 14
}

checkStock(products)
```

```python
# 3.Student Result Processing System
def checkResult(marks):
    total = 0

    #calculate total
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} : Pass"
    else:
        return f"Average: {average:.2f} : Fail"


studentMarks = [70, 35, 60, 55]
print(checkResult(studentMarks))
```

```python
# 4.Cab Fare Estimator with Retry Option
def calculateFare(distance,peak):
    baseFare = 50
    fare = baseFare + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25
    return fare

# Retry
while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    totalFare = calculateFare(distance, peak)
    print(f"Total Fare : {totalFare:.2f}")
    break
    print("Invalid input. Please try again.")
```

```python
# 5.Employee Attendance Eligibility Checker
def checkAttendance(attendanceList):
    present_days = 0

    for day in attendanceList:
        if day.lower() == "p":
            present_days += 1

    percentage = (present_days / len(attendanceList)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% : Eligible"
    else:
        return f"Attendance: {percentage:.2f}% : Not Eligible"

attendance = ["P", "A", "P", "P", "P", "A", "P"]
print(checkAttendance(attendance))
```

```python
# 6.Password Strength Checker
def checkPassword(password):
    digit = False
    special = False

    if len(password) < 8:
        return "Weak Password : Minimum 8 characters required"

    for char in password:
        if char.isdigit():
            digit = True
        if char in "@#$":
            special = True

    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password : Must contain digit & special character"

userPassword = input("Enter password: ")
print(checkPassword(userPassword))
```
