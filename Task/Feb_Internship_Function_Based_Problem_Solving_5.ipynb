```python
# 1.Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


n = int(input("Enter number of posts: "))
likes_list = []
for i in range(n):
    likes = int(input(f"Enter likes for post {i+1}: "))
    likes_list.append(likes)

analyze_post_engagement(likes_list)
```

```python
# 2.Healthcare – Medicine Stock Alert System

def check_medicine_stock(stock):

    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


stock = int(input("Enter medicine stock quantity :"))
check_medicine_stock(stock)
```

```python
# 3.Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_data,required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

n = int(input("Enter number of rainfall readings: "))

rainfall_data = []
for i in range(n):
  rain = float(input(f"Enter rainfall {i+1}: "))
  rainfall_data.append(rain)

required_level = float(input("Enter required rainfall level: "))
rainfall_checker(rainfall_data,required_level)
```

```python
# 4.Social Media – Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

n = int(input("Enter number of usernames: "))

usernames = []
for i in range(n):
    username = input(f"Enter username {i+1}: ")
    usernames.append(username)
detect_duplicate_accounts(usernames)
```

```python
# 5.Healthcare – Appointment Eligibility Checker
def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

age = int(input("Enter patient age: "))
check_appointment_eligibility(age)
```

```python
# 6.Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

n = int(input("Enter number of crop prices: "))

prices = []
for i in range(n):
    price = int(input(f"Enter price {i+1}: "))
    prices.append(price)

filter_premium_crops(prices)
```

```python
# 7.System Monitoring – Application Health Checker
def check_application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

errors = int(input("Enter error count: "))
check_application_health(errors)
```

```python
# 8.Banking – Daily Transaction Limit Checker
def check_transaction(amount,limit):
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

amount = int(input("Enter transaction amount: "))
limit = 50000

check_transaction(amount,limit)
```

```python
# 9.E-Learning – Student Attendance Eligibility
def check_exam_eligibility(percentage):
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

total_classes = int(input("Enter total number of classes: "))
attended = int(input("Enter number of classes attended: "))

percentage = (attended / total_classes) * 100

check_exam_eligibility(percentage)
```

```python
# 10.Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Level:", usage)

units = int(input("Enter units consumed: "))

calculate_electricity_bill(units)
```
