```python
# 1.Unique Words in a Sentence

#input sentence
Sentence = input("Enter a Sentence :")

#split sentence into words
words=Sentence.split()

#To get unique words convert list to set
words=set(words)

#Result
print("\nUnique Words Count :",len(words))
print("Unique Words :",words)
```

```python
# 2.Highest Salary from Employee Data by using for loop
employees={"Ravi": 75000,"Anita": 68000,"Kiran": 72000}

HighSalaryEmployee=""
HighSalary=0

for name,salary in employees.items():
    if salary>HighSalary:
        HighSalary=salary
        HighSalaryEmployee=name

print(f"Highest Salary {HighSalaryEmployee} - {HighSalary}")
```

```python
# 2.Highest Salary from Employee Data By using max() and .get
employees={"Ravi": 75000,"Anita": 68000,"Kiran": 72000}

highest_employee = max(employees, key=employees.get)
#max() - for getting the key of highest salary
#employees.get used to compare values(Salaries)

print(f"Highest Salary {highest_employee} - {employees[highest_employee]}")
```

```python
# 3.Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]

#sort list for getting ascending order
SortedList=sorted(numbers)

print("List :",numbers)

#last element is Max
print("Max :",SortedList[-1])

#first element is Min
print("Min :",SortedList[0])
```

```python
# 3.Find Maximum and Minimum Values using loop
numbers = [45, 22, 89, 10, 66]

max = numbers[0]
min = numbers[0]

for num in numbers:
    if num>max:
        max=num
    if num<min:
        min=num

print("List :", numbers)
print("Max :", max)
print("Min :", min)
```

```python
# 4.Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
cnt=0

for i in prices:
  if i>1000:
    cnt+=1

print("Product above 1000 :",cnt)
```

```python
# 5.Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
cnt=0

for i in attendance:
  if i=="P":
    cnt+=1

percentage=(cnt/len(attendance))*100

print("Attendance Percentage :",percentage)
```

```python
# 6.Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

UniqueNumbers=set(phone_numbers)

print("Unique phone numbers :",UniqueNumbers)
```

```python
# 7.Count Character Frequency
StrInput = input("Enter a String :")

charCount={}

for i in StrInput:
  if i in charCount:
    charCount[i]+=1
  else:
    charCount[i]=1

print(charCount)
```

```python
# 10.Calculate Average Marks
marks=[85, 90, 78, 92, 88,55,80]

averageMarks=sum(marks)/len(marks)

#round() is for rounds the number of 2 decimal places
#round converts 81.14285714285714 to 81.14
print("Average Marks :",round(averageMarks,2))
```
