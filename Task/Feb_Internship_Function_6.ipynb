```python
# 1.Smart Parking Lot Management System
def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "in":
            parked += 1
        elif log == "out" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

capacity = 50
logs = []

number_of_logs = int(input("Enter the number of logs :"))

for i in range(number_of_logs):
  log_value = input(f"Enter value for log {i+1} :")
  logs.append(log_value)

smart_parking(capacity, logs)
```

```python
# 2.Online Food Delivery Time Estimator
def estimate_delivery(distance, traffic, weather):
    time = distance * 5

    if traffic == "high":
        time += 15
    elif traffic == "medium":
        time += 5

    if weather == "rainy":
        time += 10
    elif weather == "storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")

distance = int(input("Enter the distance in km :"))
traffic = input("Enter the traffic condition (low, medium, high) :").lower()
weather = input("Enter the weather (rainy,storm):").lower()

estimate_delivery(distance, traffic, weather)
```

```python
# 3.Movie Theatre Seat Occupancy Analyzer
def theatre_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show's Status: Housefull")
    elif occupancy >= 75:
        print("Show's Status: Almost Full")
    else:
        print("Show's Status: Seats Available")

total_seats = 200

booked = int(input("Enter number of booked seats :"))

booked_seats = [1] * booked
theatre_occupancy(total_seats, booked_seats)
```

```python
# 4.Cloud Server Load Classification System
def server_load(CPU_readings):
    avg = sum(CPU_readings) / len(CPU_readings)
    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


CPU_readings = []
No_of_readings = int(input("Enter the number of CPU readings :"))

for i in range(No_of_readings):
  reading_value = int(input(f"Enter value for reading {i+1} :"))
  CPU_readings.append(reading_value)

server_load(CPU_readings)
```

```python
# 5.Smart Classroom Resource Usage Monitor
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused))

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")

projector = int(input("ENter no of projectors :"))
AC = int(input("ENter no of AC :"))
Lights = int(input("ENter no of Lights :"))

resources = {
    "Projector": projector,
    "AC": AC,
    "Lights": Lights
}

resource_monitor(resources)
```

```python
# 6.Online Event Registration Capacity Controller
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

capacity = 100
registrations = int(input("Enter the number of registrations :"))
event_registration(capacity, registrations)
```
